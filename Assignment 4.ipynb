### Questions

1. What command would you use to display all the commits that have not yet been pushed to the remote repository?
   
2. Explain the purpose of the `git tag` command.
   
3. How do you remove a file from the repository history?
   
4. How do you temporarily save your changes without committing them?
   
5. How do you amend the most recent commit message?
   
6. How can you split a commit into multiple commits?
   
7. How do you display the changes between two branches?
   
8. What command would you use to see which remote repository a local repository is connected to?
   
9.  Explain the difference between `git merge` and `git rebase`.
    
10. How can you search for a specific commit message?

1.
I would use git log origin/main..HEAD
This shows commits that are on my local branch but have not been pushed to the remote repository yet.

2.
git tag is used to give a name to a specific commit, usually to mark a version or release.

3.
git filter-repo removes the file from the entire Git history.

4.
git stash is used to save unfinished work and clear my working directory for later.

5.
git commit --amend
This is used to edit the message of my last commit.

6.
git rebase -i HEAD~1
Interactive rebase allows me to stop at a commit, break its changes apart, and recommit them as separate commits.

7.
git diff
This command shows what is different between two branches (after the diff, I would add the two branch names).

8.
git remote -v
This shows the remote repository URLs of where the project is connected to.

9.
git merge combines two branches and keeps the commit history.
git rebase moves commits onto another branch to create a cleaner history.

10.
git log --grep=
After the =, I would put the specific message I am looking for to find the commit containing that text.

### Question 04

Convert the hexadecimal colour #2A9F3B to its RGB components. Show your steps.

```python
hexadecimal_color = "#2A9F3B"
hexadecimal_color = hexadecimal_color.lstrip('#')

red = int(hexadecimal_color[0:2], 16)
green = int(hexadecimal_color[2:4], 16)
blue = int(hexadecimal_color[4:6], 16)

print(f"rgb({red},{green},{blue})")
```

### Question 05

Using the coin representation system described in the lecture (c[quarters][dimes][nickels][pennies]), convert $1.37 to coin representation. Explain your reasoning.

```python
amount = 1.37

cents = int(amount * 100)

quarters = cents // 25
cents = cents % 25

dimes = cents // 10
cents = cents % 10

nickels = cents // 5
cents = cents % 5

pennies = cents

print("c" + str(quarters) + str(dimes) + str(nickels) + str(pennies))
```
